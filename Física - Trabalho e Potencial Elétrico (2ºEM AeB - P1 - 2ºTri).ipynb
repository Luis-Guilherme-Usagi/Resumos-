# Trabalho da força elétrica

## Trabalho da força elétrica em um campo elétrico uniforme

  Se uma determinada carga encontrar-se sob influência de apenas uma região onde atuam forças elétricas, existirá um deslocamento gerado por essa força. Desta forma, estará em plena realização de trabalho, pois tem a capacidade de transformar energia potencial elétrica em energia cinética ou o inverso.
  Este trabalho pode ser descrito pela expressão:
  
  $$ Τrabalho = Força_e . Distância $$
  
  Sendo o trabalho expresso por:
  
  $$ Força_e = |Carga| . Campo Elétrico $$
  
  Desta forma, o trabalho de uma força elétrica pode ser expresso por:
  
  * $ Trabalho = |Carga|.Campo Elétrico.Distância $  
  
  **ou**  
  
  * $ T=|q|.E.d $
  
  E como $ΔPotencial Elétrico = Campo Elétrico . Distância$, ou seja, $U=E.d$ outra representação possível desta fórmula é:
  
  * $T=q.U$
  
________________________________________________________________________________________________________________________________
  **Nota:** As grandezas, carga, trabalho e diferença de potencial são dados respectivamente em coulomb, joule e volts.
________________________________________________________________________________________________________________________________
  
  ### Conclusões
  
  * Todo deslocamento perpendicular a linha de campo denota trabalho nulo.
  
  * Cargas positivas se deslocam de pontos de maior para pontos de menor potencial, enquanto as negativas deslocam-se de pontos de menor potencial para pontos de maior potencial.
  
  * Deslocamentos diagonais em relação as linhas de campo são expressos por: $ Trabalho=Força.Distância.Cosα $

## Trabalho da força elétrica em um campo elétrico qualquer e diferença de potencial elétrico.

  Ao mover uma carga entre dois pontos distintos de um campo elétrico qualquer, haverá deslocamento entre dois pontos de potenciais diferentes que determinam uma d.d.p ou tensão elétrica, representada por $V_A-V_B$ que é simbolizada por U.
  
$V_A-V_B= \frac {TrabalhoAB}{Carga}$ implica na expressão $U= \frac {TrabalhoAB}{Carga}$

No qual $TrabalhoAB=Carga.(V_A-V_B)$

# Diferença de potencial (d.d.p)

## Potencial elétrico no campo de uma carga puntiforme

  O potencial elétrico em um ponto da região do campo elétrico é definido pela razão entre a medida do trabalho da força elétrica efetuado sobre uma carga de prova e a quantidade de carga elétrica deslocada. Ao se colocar uma carga de prova **q** sobre um ponto P da região do campo elétrico, ela adquire uma energia associada ao quão predisposta está a entrar em movimento.
  
  Define-se por $V$ o potencial elétrico associado a uma carga, tendo, assim:
  
  * $V= \frac{Trabalho}{Carga}$
  
  
  
  * $V= \frac{Força_e.Distância}{Carga}$
  
  
  
  * $V= \frac { \frac {Constante_e.CargaPuntiforme.Carga.Distância}{Distância^2}}{Carga}$
  
  
  
  * $V = \frac {Constante_e.CargaPuntiforme}{Distância}$
  
  Onde:
  
 * $V$=Potencial elétrico **(volt)**
 
 * $K$ **ou** $Constante_e$=Constante eletrostática **(no vácuo $ 9.10^9\frac {N.m^2}{c^2}$ ou $ 9.10^9\frac {V.m}{c}$ )**
 
 * $Q$=Carga puntiforme **(coulomb)**
 
 * $d$=Distância **(metros)**
 
 * $T$=Trabalho **(joule)**
 
 _______________________________________________________________________________________________________________________________
  
  ## Propriedade do potencial elétrico
  
  Grandeza escalar, o potencial elétrico pode ser definido pela razão entre outras grandezas escalares: energia e carga elétrica.
  
  * $Carga Puntiforme>0$ **implica em** $Potencial Elétrico>0$
  
  * $Carga Puntiforme<0$ **implica em** $Potencial Elétrico<0$
  
  * $Carga Puntiforme=0$ **implica em $Potencial Elétrico=0$
  
  O potencial elétrico de um ponto independe da carga elétrica de prova **q**,que pode ser colocada sobre esse ponto, uma vez que o potencial elétrico é uma propriedade do campo.
  
  ________________________________________________________________________________________________________________________________
  
  ## Potencial elétrico resultante gerado por várias cargas elétricas
  
  Por ser uma grandeza escalar,a sujeição de um único ponto a **n** pontos pode ser representada por sua soma algébrica, ou seja:
  
  * $PotencialElétrico_P=PotencialElétrico_1+PotencialElétrico_2+PotencialElétrico_3...+PotencialElétrico_n$
  
  **ou**
  
  * $PotencialElétrico_P=\frac {CargaPuntiforme_1}{Distância_1}+\frac {CargaPuntiforme_2}{Distância_2}+\frac {CargaPuntiforme_3}{Distância_3}+\frac {CargaPuntiforme_n}{Distância_n}$
   ________________________________________________________________________________________________________________________________
   
   ## Diferença de potencial (d.d.p.) ou tensão elétrica
   
   A relação da diferença de potencial entre dois pontos da região de um campo elétrico, $PotencialElétrico_A-PotencialElétrico_B=\frac{TrabalhoAB}{Carga}$, permite que seja determinado o potencial elétrico de um ponto, para isso, é necessário estabelecer que um dos pontos seja o potencial de referência. Sendo assim, seu valor será considerado nulo $(PotencialElétrico_B=0)$. Logo:
   
   $V_A-V_B= \frac{TrabalhoAB}{Carga}$**, sendo $V_B=0$, tem-se** $V_A= \frac {TrabalhoAB}{Carga}$
