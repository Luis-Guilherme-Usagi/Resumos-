# Leis de Kepler 

### Primeira Lei de Kepler - Lei das Órbitas

Os planetas descrevem órbitas elípticas em torno do sol, que fica em um dos focos desta elipse.

### Segunda Lei de Kepler - Lei das Áreas

O segmento que une o centro do sol e um planeta varre áreas iguais em intervalos de tempos iguais.

**Implicações:**

* Pela Segunda Lei de Kepler, se as áreas forem iguais, o intervalo de tempo para o planeta descrever um arco é igual ao tempo para percorrer o outro.

* Pode-se perceber que os planetas não mantêm a uma velocidade constante em torno do sol, pois para que o planeta percorra duas áreas distintas numa elipse, sendo elas referentes a um dos focos, em tempos iguais, é necessário que a velocidade do arco de maior comprimento seja maior.

### Terceira Lei de Kepler - Lei dos Períodos

$$\frac {Período^2}{Raio Médio^3} = K $$

Sendo que o **Raio Médio** pode ser obtido a partir de:

$$ RaioMédio = \frac {d_A + d_B} {2} $$

**Implicações:**

Analisando esta lei, nota-se que, quanto maior a distância de um planeta em relação ao sol, maior será seu período, ou seja, o tempo necessário para translação.

### Lei da Gravitação Universal

$$ F = Constante Gravitacional \frac {MassaMaior . Massa Menor }{d^2} $$

# Equilíbrio de um ponto material

** Em caso de vetores diagonais, a decomposição é feita da seguinte forma:**

$$ v_x = v . cos  α $$ 

$$ v_y = v . sen α $$

**Nota:** Para um ponto material estar em equilíbrio estático, as somas de suas projeções sobre os eixos perpendiculares e coplanares devem ser nulas. **(ΣF = 0)**

### Polia

**Polia Fixa -** Modifica apenas a direção e sentido da força, não altera sua intensidade.

** Polia Móvel -** Modifica a direção, sentido e intensidade da força, sendo que, cada polia móvel do sistema reduz a força necessária à metade.

### Triângulos Egípcios

** 45º -** Os catetos equivalem a $ 1 $ e a hipotenusa $ √2 $. ** OU** Os catetos equivalem a $ √2 $ e a hipotenusa $ 2 $.

** 30º -** O cateto oposto vale $ \frac {1}{2} $, o adjacente vale $ \frac {√3}{2} $ e a hipotenusa $ \frac {√3}{3} $.

** 60º -** O cateto oposto vale $ \frac {√3}{2} $, o adjacente vale $ \frac {1}{2} $ e a hipotenusa $ √3 $

** 53º e 37º -** O triângulo é pitagórico.

# Equilíbrio de um corpo extenso

### Momento binário

O momento é negativo se possui sentido anti-horário, e binário de possui sentido horário.

### Tipos de Alavanca

* **Interfixa -** O ponto **O**, ou **Ponto de Apoio** se localiza entre o **Braço Potente** e o **Braço Resistente**.

* **Interpotente -** A resitência se localiza entre o **Braço Potente** e o **Ponto de Apoio**.

* ** Inter-resistente -** A potência se localiza entre o **Braço Resistente** e o **Ponto de Apoio**.
